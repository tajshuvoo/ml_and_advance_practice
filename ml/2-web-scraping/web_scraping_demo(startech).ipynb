```python
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://www.startech.com.bd/component?page={}"
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response_content = "" 

for page_num in range(1, 205): 
    url = BASE_URL.format(page_num)
    print(f"Fetching page {page_num}...")
    resp = requests.get(url, headers=HEADERS)
    resp.raise_for_status()
    
    
    response_content += resp.text
    
    time.sleep(2) 
```

```python
soup = BeautifulSoup(response_content, "html.parser")
```

```python
products= soup.find_all('div',class_='p-item-details')
```

```python
products[0]
```

```python
product_list = []

for product in products:
    # Extract product title
    name_tag = product.find('h4', class_='p-item-name')
    title = name_tag.get_text(strip=True) if name_tag else 'N/A'

    # Extract product link from <a> inside <h4>
    link_tag = name_tag.find('a') if name_tag else None
    link = link_tag.get('href') if link_tag and link_tag.has_attr('href') else 'N/A'

    # Extract short description features
    desc_div = product.find('div', class_='short-description')
    features = [li.get_text(strip=True) for li in desc_div.find_all('li')] if desc_div else []

    # Find the price container
    price_div = product.find('div', class_='p-item-price')

    # Default to 'N/A'
    new_price = old_price = 'N/A'

    if price_div:
        # Try to find new and old price spans
        new_price_tag = price_div.find('span', class_='price-new')
        old_price_tag = price_div.find('span', class_='price-old')

        if new_price_tag:
            new_price = new_price_tag.get_text(strip=True)
        elif price_div.get_text(strip=True):  
            new_price = price_div.get_text(strip=True)

        if old_price_tag:
            old_price = old_price_tag.get_text(strip=True)

    # Save all data into a dictionary
    product_data = {
        'title': title,
        'link': link,
        'features': features,
        'new_price': new_price,
        'old_price': old_price
    }

    # Add to the list
    product_list.append(product_data)
```

```python
df=pd.DataFrame(product_list)
```
